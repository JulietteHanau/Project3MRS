```python
import pandas as pd
```

```python
# Read the table
final_table = pd.read_csv('https://raw.githubusercontent.com/Sebastiao199/Project3MRS/main/4_LinkingTables/tprn_10.csv')
final_table
```

```python
# Filter per 'director' category

condition = final_table[final_table['category'] == 'director']
condition
```

```python
# Dropped not necessary columns

condition.drop(['Year', 'Genres'], axis = 1, inplace = True) 
```

```python
# Sorted by directors name 
condition.sort_values(by=["Staff name"]) 
```

```python
# Grouping by directors and counting their number of movies 
condition2 = condition.groupby('Staff name').agg({'Movie title':'count'})
```

```python
condition2
```

# PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sorting by number of movies done per directors
final = condition2.sort_values(by='Movie title', ascending=False).reset_index()
```

```python
ten_director = final.head(10)
```

```python
five_director = final.head()
```

```python
five_director
```

```python
ten_director
```

```python
sns.barplot(x = 'Staff name', y = 'Movie title', data = ten_director)

#sns.set_xticklabels(sns.get_xticklabels(), rotation = 45, horizontalalignment = 'center')
plt.xticks(rotation=40)
```

```python
sns.barplot(x = 'Staff name', y = 'Movie title', data = five_director, palette="mako")

#sns.color_palette("mako", as_cmap=True)

#sns.set_xticklabels(sns.get_xticklabels(), rotation = 45, horizontalalignment = 'center')
plt.xticks(rotation=50)
#plt.xticks(rotation ='horizontal')

plt.title('Top 5 directors with the greatest number of films')

plt.xlabel('Directors name')
plt.ylabel('Number of movies')


```
