```python
#scaler = preprocessing.StandardScaler()
scaler_minmax = preprocessing.MinMaxScaler()
df_to_standard = df_ml_100_vect[['startYear_x', 'runtimeMinutes_x', 'averageRating', 'numVotes']]
#standard_df = scaler.fit_transform(df_to_standard)
standard_df = scaler_minmax.fit_transform(df_to_standard)
standard_df = pd.DataFrame(standard_df, columns = ['startYear_x', 'runtimeMinutes_x', 'averageRating', 'numVotes'])
```

```python
wordnet_lemmatizer = WordNetLemmatizer()

#def lemm(text):
 #   text = wordnet_lemmatizer.lemmatize(text)
  #  return text

def cleaning(text):
    # turn into list
    tokens = nltk.word_tokenize(text)
    # tokens_lower = list(map(lambda t: t.lower(), tokens))
    #tokens = list(filter(remove_stopwords, tokens))  
    #tokens = list(map(lemm, tokens))
    return " ".join(tokens)

df_after_standard['genre_clean'] = df_after_standard['genres_x'].apply(cleaning)
df_after_standard#.reset_index(inplace=True)
```

```python
#df_ml_100_vect=df_ml_100_vect.tail(n=90000)
#df_ml_100_vect=df_ml_100_vect.head(139000)
```

```python
cv = CountVectorizer()
GenreX = cv.fit_transform(df_after_standard["genre_clean"]).toarray()

print("Count Vector : \n",GenreX)
print("\nNote: First row of above count vector: ",GenreX[0])
print("\nColumns Coresponding to above count vector is :\n",cv.get_feature_names_out())
print(GenreX.shape)
X_df=pd.DataFrame(GenreX, columns=['action', 'adult' ,'adventure' ,'animation' ,'biography' ,'comedy', 'crime', 'documentary', 'drama', 'family', 'fantasy', 'fi', 'game', 'history', 'horror', 'music' ,'musical' ,'mystery' ,'news' ,'reality' ,'romance', 'sci', 'short', 'show', 'sport', 'talk', 'thriller', 'tv' ,'war', 'western'])

output = pd.concat([df_after_standard, X_df], join='inner', ignore_index=False, axis=1)
output
```
